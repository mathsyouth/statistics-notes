```python
%matplotlib inline

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
```

The following code snippets are implemented to reproduce the results of the blog [UNDERSTANDING P-VALUES VIA SIMULATIONS](https://researchutopia.wordpress.com/2013/11/10/understanding-p-values-via-simulations/). First, we simulate p-values distribution under the assumption that the null is true.

```python
# Set the number of simulated experiments
num_sims = 100000
# Set the zero array to store computed p-values
p_array = np.zeros(num_sims)
mu = 100
sigma = 20
size = 100
for i in range(num_sims):
    x = np.random.normal(loc=mu, scale=sigma, size=size)
    y = np.random.normal(loc=mu, scale=sigma, size=size)
    _, p_array[i] = stats.ttest_ind(x, y)
```

```python
plt.hist(p_array, bins=100)
plt.title("Histogram of p-values under the null")
plt.xlabel("Observed p-value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Second, we simulate p-values distribution under the assumption that the null is false.

```python
# Set the number of simulated experiments
num_sims = 100000
# Set the zero array to store computed p-values
p_array = np.zeros(num_sims)
mu = 100
sigma = 20
size = 100
for i in range(num_sims):
    x = np.random.normal(loc=mu+3, scale=sigma, size=size)
    y = np.random.normal(loc=mu, scale=sigma, size=size)
    _, p_array[i] = stats.ttest_ind(x, y)
```

```python
plt.hist(p_array, bins=50)
plt.title("Histogram of p-values (true group difference)")
plt.xlabel("Observed p-value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
